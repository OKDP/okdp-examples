# Gold Mobility Index

This notebook is the navigation index for the Gold mobility area.

It explains the purpose of the Gold mobility notebooks, the current available notebooks, and the recommended execution path.

## Gold layer purpose

The Gold layer publishes business-ready, consumer-facing datasets built only from trusted Silver tables.

For mobility, Gold notebooks should convert the detailed Silver trip table into curated aggregates and reporting-friendly datasets.


## Scope of this folder

This folder covers the Gold mobility domain.

Current domain implemented here:

- `nyc_trip`

Current Gold data product available here:

- `gold.nyc_tlc.monthly_stats`


## Available notebooks

### Monthly business aggregate

- [nyc_trip/01_build_monthly_stats.ipynb](nyc_trip/01_build_monthly_stats.ipynb)

Purpose:

- reads `silver.nyc_tlc.trips`
- aggregates the trusted trip-level Silver dataset to monthly business statistics
- publishes `gold.nyc_tlc.monthly_stats`


## Recommended usage

Use the Gold notebook only after the Silver table `silver.nyc_tlc.trips` has been successfully built and validated.

Gold notebooks must not bypass Silver and read directly from Bronze.

## Runtime notes

- Spark, Polaris, Iceberg, and object-storage access are configured outside the notebooks
- Sensitive Spark configuration is intentionally hidden from notebook code

## Current environment caveat

In the current environment, SeaweedFS may fail during Iceberg multipart upload initialization.

The typical failure is `CreateMultipartUpload` with `403 Access Denied`.

For that reason, Gold notebooks in this showcase are intentionally kept simple and conservative on the write path.
