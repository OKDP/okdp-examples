# NYC Tripdata Lakehouse Jupyter notebooks

Welcome to the Gold Mobility Workspace!  
This area contains the notebooks used to **publish curated analytical outputs** from the trusted Silver mobility table in the NYC Tripdata lakehouse showcase.

These notebooks focus on the **consumer-facing layer** of the project, where detailed Silver trip data is converted into business-ready aggregates and reporting-friendly datasets.

In this area, you will find notebooks covering topics such as:

- Gold aggregations derived from trusted Silver tables only  
- Consumer-oriented analytical datasets  
- Monthly business statistics and reporting-friendly outputs  
- Future expansion toward additional Gold mobility data products  

The Gold layer is where the project exposes stable analytical products that are simpler for downstream consumers than the detailed Silver trip table.

Use this index as your starting point to navigate the Gold mobility notebooks and understand the current implemented Gold outputs and the planned expansion path.

## Runtime notes

- Spark, Polaris, Iceberg, and object-storage access are configured outside the notebooks  
- Sensitive Spark configuration is intentionally hidden from notebook code  

## Current environment caveat

In the current environment, SeaweedFS may fail during Iceberg multipart upload initialization.

The typical failure is `CreateMultipartUpload` with `403 Access Denied`.

For that reason, Gold notebooks in this showcase are intentionally kept simple and conservative on the write path.
