# Silver Mobility Index

This notebook is the navigation index for the Silver mobility area.

It explains the role of the Silver mobility notebooks, the current available notebooks, and the recommended execution path.

## Silver layer purpose

The Silver layer publishes trusted, conformed analytical datasets built from Bronze raw data.

For mobility, the main Silver objective is to publish the canonical trip-level table used by downstream Gold notebooks.


## Scope of this folder

This folder covers the Silver mobility domain.

Current domain implemented here:

- `nyc_trip`

Primary dataset produced by this area:

- `silver.nyc_tlc.trips`


## Available notebooks

### Standard build notebook

- [nyc_trip/01_build_trips.ipynb](nyc_trip/01_build_trips.ipynb)

Purpose:

- reads Bronze raw Yellow and Green trip files
- maps them to a canonical schema
- applies Silver filtering and validation logic
- publishes `silver.nyc_tlc.trips`

### Environment-specific troubleshooting notebook

- [nyc_trip/01_Issue-seaweedFS-build_trips.ipynb](nyc_trip/01_Issue-seaweedFS-build_trips.ipynb)

Purpose:

- documents or tests the Silver build path in the current SeaweedFS-constrained environment
- should be used only when the standard Silver notebook needs an environment-specific workaround or comparison


## Recommended usage

### Default recommendation

Use `nyc_trip/01_build_trips.ipynb` as the main Silver notebook.

### When to use the issue notebook

Use `nyc_trip/01_Issue-seaweedFS-build_trips.ipynb` only if you are troubleshooting the current SeaweedFS write-path behavior or comparing notebook variants.

The issue notebook should not be treated as the default production path unless the environment specifically requires it.


## Silver contract summary

The expected Silver data product in this area is `silver.nyc_tlc.trips`.

### Role of the table

- trusted trip-level conformance layer
- stable analytical input for Gold notebooks
- unified schema across Yellow and Green trip sources

### Downstream dependency

Gold mobility notebooks must read from `silver.nyc_tlc.trips`, not directly from Bronze.


## Runtime notes

- Spark, Polaris, Iceberg, and object-storage access are configured outside the notebooks
- Sensitive Spark configuration is intentionally hidden from the notebook code

## Current environment caveat

In the current environment, SeaweedFS may fail during Iceberg multipart upload initialization.

The typical failure is `CreateMultipartUpload` with `403 Access Denied`.

This is a storage write-path issue and should not be confused with the Silver transformation design itself.
