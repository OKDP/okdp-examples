# NYC Tripdata Lakehouse Jupyter notebooks

Welcome to the Silver Mobility Workspace!  
This area contains the notebooks used to **build, validate, and troubleshoot** the Silver mobility data product for the NYC Tripdata lakehouse showcase.

These notebooks focus on the **trusted conformance layer** of the project, where Bronze raw data is transformed into a canonical analytical table.

In this area, you will find notebooks covering topics such as:

- Canonical schema mapping between Yellow and Green trip datasets  
- Silver validation and filtering logic  
- Trusted Iceberg table publishing in Polaris  
- Environment-specific troubleshooting for the Silver write path  

The Silver layer is where the project publishes the detailed trip-level dataset used by downstream Gold notebooks.

Use this index as your starting point to navigate the Silver mobility notebooks and understand which notebook should be used in the standard path versus troubleshooting scenarios.


## Silver mobility notebooks

The following notebooks belong to the Silver mobility area and focus on publishing the trusted canonical trip table for downstream analytical use.

### 1. Standard Silver build notebook
- **Notebook:** `nyc_trip/01_build_trips.ipynb`
- **Description:** The main Silver build workflow. It reads Bronze Yellow and Green raw trip files, maps them to a canonical schema, applies validation and filtering logic, and publishes the conformed Silver table.
- ➡️ **Open:** [Build Silver trips](./nyc_trip/01_build_trips.ipynb)

### 2. Environment-specific troubleshooting notebook
- **Notebook:** `nyc_trip/01_Issue-seaweedFS-build_trips.ipynb`
- **Description:** A troubleshooting and comparison notebook kept for the current SeaweedFS-constrained environment. It should only be used when investigating or documenting environment-specific write-path issues.
- ➡️ **Open:** [SeaweedFS issue notebook](./nyc_trip/01_Issue-seaweedFS-build_trips.ipynb)


## Silver data product summary

The primary dataset produced by this area is:

- `silver.nyc_tlc.trips`

This table is the trusted Silver contract for the mobility domain and acts as:

- the canonical trip-level conformance layer  
- the stable analytical input for Gold notebooks  
- the unified schema across Yellow and Green trip sources  

Gold mobility notebooks are expected to read from `silver.nyc_tlc.trips`, not directly from Bronze.


## Recommended usage

Use the notebooks in this order for the standard Silver path:

1. Review the Bronze notebooks to understand source schemas and data-quality observations  
2. Run `nyc_trip/01_build_trips.ipynb` to publish the canonical Silver table  
3. Continue to the Gold mobility index once the Silver table has been validated  

### Default recommendation

Use `nyc_trip/01_build_trips.ipynb` as the main Silver notebook.

### When to use the issue notebook

Use `nyc_trip/01_Issue-seaweedFS-build_trips.ipynb` only when you are troubleshooting the current SeaweedFS write-path behavior or comparing notebook variants.

## Runtime notes

- Spark, Polaris, Iceberg, and object-storage access are configured outside the notebooks  
- Sensitive Spark configuration is intentionally hidden from notebook code  

## Current environment caveat

In the current environment, SeaweedFS may fail during Iceberg multipart upload initialization.

The typical failure is `CreateMultipartUpload` with `403 Access Denied`.

This is a storage write-path issue and should not be confused with the Silver transformation design itself.
